```python
import pandas as pd
import nltk 
```

```python
df= pd.read_csv('/content/train.csv')
```

```python
df.head()
```

```python
# droping NAN values
df = df.dropna()
```

```python
## getting independent feature
X = df.drop('label',axis=1)
```

```python
y = df['label']
```

```python
import tensorflow as tf
```

```python
y.shape
```

```python
from tensorflow.keras.layers import Embedding 
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size
voc_size = 5000
```

<h1> One_hot_representation </h1>


```python
messages = X.copy()
```

```python
messages.reset_index(inplace = True)
```

```python
import nltk 
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
nltk.download('stopwords')
nltk.download()
```

```python
### Dataset Preprocessing
ps = PorterStemmer()
lm = WordNetLemmatizer()
corpus = []
for i in range(0,len(messages)):
  review = re.sub('[^a-zA-Z]'," ",messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [lm.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from tensorflow.keras.preprocessing.text import one_hot
one_hot_rep =[one_hot(words,voc_size) for words in corpus]
one_hot_rep
```

<h1> Embedding Representation<h1>

```python
sent_length = 20
embedded_docs = pad_sequences(one_hot_rep,padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## creating model
embedding_vector_features = 50
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.3,random_state=42)
```

```python
## Model training
model.fit(X_train,y_train,validation_data =(X_test,y_test),epochs=20,batch_size=64)
```

<h1> Performance Metrics and Accuracy
</h1>

```python
y_pred = model.predict(X_test)
prediction_result = []
for i in y_pred:
  if i[0] < 0.5:
    prediction_result.append(0)
  else:
    prediction_result.append(1)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction_result)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction_result)
```
